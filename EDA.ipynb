```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'Internet__.xlsx'
xls = pd.ExcelFile(file_path)

hojas = xls.sheet_names
print(f"Hojas en el archivo: {hojas}")

def eda_completo(df, nombre_hoja):
    print(f"\n\n--- EDA para la hoja: {nombre_hoja} ---")
    
    # Mostrar las primeras filas
    print("\nPrimeras filas:")
    display(df.head())
    
    # Descripción estadística básica
    print("\nDescripción estadística:")
    display(df.describe())
    
    # Información de las columnas y tipos de datos
    print("\nInformación del DataFrame:")
    display(df.info())
    
    # Búsqueda de valores faltantes
    print("\nValores nulos por columna:")
    display(df.isnull().sum())
    
    # Detección de duplicados
    print("\nNúmero de registros duplicados:")
    display(df.duplicated().sum())
    
    # Gráficos para variables numéricas
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
    if not numeric_cols.empty:
        print("\n--- Distribución de las columnas numéricas ---")
        
        # Histograma
        df[numeric_cols].hist(figsize=(12, 10), bins=15)
        plt.suptitle(f'Distribución de columnas numéricas - {nombre_hoja}')
        plt.show()

        # Boxplot para detectar outliers
        plt.figure(figsize=(12, 8))
        df[numeric_cols].boxplot()
        plt.title(f'Outliers en columnas numéricas - {nombre_hoja}')
        plt.xticks(rotation=90)
        plt.show()
    
    # Gráficos para variables categóricas
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns
    if not categorical_cols.empty:
        print("\n--- Distribución de las columnas categóricas ---")
        
        for col in categorical_cols:
            plt.figure(figsize=(12, 8))
            df[col].value_counts().plot(kind='bar')
            plt.title(f'Distribución de {col} - {nombre_hoja}')
            plt.xticks(rotation=90)
            plt.show()

for hoja in hojas:
    # Cargar cada hoja
    df_hoja = pd.read_excel(xls, sheet_name=hoja)
    
    # Realizar el EDA para cada hoja
    eda_completo(df_hoja, hoja)
```
